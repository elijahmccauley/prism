```python
def safe_get(driver, url, retries=3):
    for i in range(retries):
        try:
            driver.get(url)
            return
        except Exception as e:
            print(f"Attempt {i+1} failed: {e}")
            time.sleep(3)
    raise Exception(f"Failed to load {url} after {retries} attempts.")
```

```python
def merge_athlete_times_by_grade(df5k, df3200, df1600, df800,
                                 athlete_id_cols=['AthleteName', 'Graduation_Year'],
                                 year_competed_col='Year_Competed',
                                 graduation_year_col='Graduation_Year', # Explicitly needed for calculation
                                 time_col='Time'):
    """
    Merges multiple rows of athlete data into a single row per athlete,
    with standardized columns for times from Freshman, Sophomore, Junior, and Senior years.

    Args:
        df5k (pd.DataFrame): The 5k input DataFrame.
        df3200 (pd.DataFrame): The 3200 input DataFrame.
        df1600 (pd.DataFrame): The 1600 input DataFrame.
        df800 (pd.DataFrame): The 800 input DataFrame.
        athlete_id_cols (list): Columns uniquely identifying an athlete.
        year_competed_col (str): Column with the year of competition.
        graduation_year_col (str): Column with the athlete's graduation year.
        time_col (str): Column with the athlete's time.

    Returns:
        pd.DataFrame: A new DataFrame with one row per athlete and their times
                      spread across 'Time_FR', 'Time_SO', 'Time_JR', 'Time_SR' columns.
    """

    df5k_processed = df5k.copy()
    df3200_processed = df3200.copy()
    df1600_processed = df1600.copy()
    df800_processed = df800.copy()

    grade_map = {
        3: 'FR',
        2: 'SO',
        1: 'JR',
        0: 'SR'
    }

    athlete_id_cols=['Athlete/School', 'Grade', 'State', 'Gender']
    time_col='Time'
    year_competed_col='Year'
    graduation_year_col='Grade'

    df5k_processed['NumericGrade'] = df5k_processed[graduation_year_col] - df5k_processed[year_competed_col] - 1
    df5k_processed['GradeLabel'] = df5k_processed['NumericGrade'].map(grade_map)
    df3200_processed['NumericGrade'] = df3200_processed[graduation_year_col] - df3200_processed[year_competed_col]
    df3200_processed['GradeLabel'] = df3200_processed['NumericGrade'].map(grade_map)
    df1600_processed['NumericGrade'] = df1600_processed[graduation_year_col] - df1600_processed[year_competed_col]
    df1600_processed['GradeLabel'] = df1600_processed['NumericGrade'].map(grade_map)
    df800_processed['NumericGrade'] = df800_processed[graduation_year_col] - df800_processed[year_competed_col]
    df800_processed['GradeLabel'] = df800_processed['NumericGrade'].map(grade_map)

    pivoted5k_df = df5k_processed.pivot_table(index=athlete_id_cols,
                                            columns='GradeLabel',
                                            values=time_col,
                                            aggfunc='first')
    pivoted3200_df = df3200_processed.pivot_table(index=athlete_id_cols,
                                            columns='GradeLabel',
                                            values=time_col,
                                            aggfunc='first')
    pivoted1600_df = df1600_processed.pivot_table(index=athlete_id_cols,
                                            columns='GradeLabel',
                                            values=time_col,
                                            aggfunc='first')
    pivoted800_df = df800_processed.pivot_table(index=athlete_id_cols,
                                            columns='GradeLabel',
                                            values=time_col,
                                            aggfunc='first')

    merged5k_df = pivoted5k_df.reset_index()
    merged5k_df.columns.name = None
    merged5k_df['FR_5k'] = merged5k_df['FR']
    merged5k_df['SO_5k'] = merged5k_df['SO']
    merged5k_df['JR_5k'] = merged5k_df['JR']
    merged5k_df['SR_5k'] = merged5k_df['SR']
    ordered_grade_time_cols5k = ['FR_5k', 'SO_5k', 'JR_5k', 'SR_5k']
    final_column_order5k = athlete_id_cols + ordered_grade_time_cols5k
    merged5k_df = merged5k_df[final_column_order5k]

    merged3200_df = pivoted3200_df.reset_index()
    merged3200_df.columns.name = None
    merged3200_df['FR_3200'] = merged3200_df['FR']
    merged3200_df['SO_3200'] = merged3200_df['SO']
    merged3200_df['JR_3200'] = merged3200_df['JR']
    merged3200_df['SR_3200'] = merged3200_df['SR']
    ordered_grade_time_cols3200 = ['FR_3200', 'SO_3200', 'JR_3200', 'SR_3200']
    final_column_order3200 = athlete_id_cols + ordered_grade_time_cols3200
    merged3200_df = merged3200_df[final_column_order3200]

    merged1600_df = pivoted1600_df.reset_index()
    merged1600_df.columns.name = None
    merged1600_df['FR_1600'] = merged1600_df['FR']
    merged1600_df['SO_1600'] = merged1600_df['SO']
    merged1600_df['JR_1600'] = merged1600_df['JR']
    merged1600_df['SR_1600'] = merged1600_df['SR']
    ordered_grade_time_cols1600 = ['FR_1600', 'SO_1600', 'JR_1600', 'SR_1600']
    final_column_order1600 = athlete_id_cols + ordered_grade_time_cols1600
    merged1600_df = merged1600_df[final_column_order1600]

    merged800_df = pivoted800_df.reset_index()
    merged800_df.columns.name = None
    merged800_df['FR_800'] = merged800_df['FR']
    merged800_df['SO_800'] = merged800_df['SO']
    merged800_df['JR_800'] = merged800_df['JR']
    merged800_df['SR_800'] = merged800_df['SR']
    ordered_grade_time_cols800 = ['FR_800', 'SO_800', 'JR_800', 'SR_800']
    final_column_order800 = athlete_id_cols + ordered_grade_time_cols800
    merged800_df = merged800_df[final_column_order800]

    final_df = merged5k_df.merge(merged3200_df, on=athlete_id_cols, how='outer')
    final_df = final_df.merge(merged1600_df, on=athlete_id_cols, how='outer')
    final_df = final_df.merge(merged800_df, on=athlete_id_cols, how='outer')
    return final_df
```
